# Using HOG features

# Trained on images cropped by cvlib

```python
#Imports
import pandas as pd
import os
from skimage.transform import resize
from skimage.io import imread
import numpy as np
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
import cv2
from skimage.feature import hog
import pickle5 as pickle
import dlib
from sklearn.utils import shuffle
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
```

```python
def get_eyes_array(face_img, eyes, target_size=(70,70)):
    """
    Input: 
    face_img(3D array): cropped face image (RGB)
    eye(list): eyes coordinates as (left,top,bottom,right)
    
    Output:
    eye_arr(1D array): concatenated hog features of left and right eye
    """
    eyes_arrays=[]
    for (ex1,ey1,ex2,ey2) in eyes:
        eye=face_img[ey1:ey2,ex1:ex2]
        h, w, _ = eye.shape
        
        if h>0 and w>0:
            eye_resized=cv2.resize(eye,target_size)
            eye_fd, _ = hog(eye_resized, orientations=8, pixels_per_cell=(16, 16),
                        cells_per_block=(1, 1), visualize=True, channel_axis=-1)

            eyes_arrays.append(eye_fd)
    
    if len(eyes_arrays)==0:
        eye_arr=np.concatenate((np.zeros(32),np.zeros(32)))
    elif len(eyes_arrays)==1:
        eye_arr=np.concatenate((eyes_arrays[0],np.zeros(32)))
    elif len(eyes_arrays)>=2:
        eye_arr=np.concatenate((eyes_arrays[0],eyes_arrays[1]))
    return eye_arr
```

```python
def detect_eyes(face_img, shape):
    """
    Input:
    face_img(2D array): grayscale image
    shape(object): this is the object returned by dlib.shape_predictor
                    this object contains 68 facial landmarks             
    Output:
    eyes(list): tuple of left and right eye coordinates in format of (left,top,right,bottom)
    """
    left = [36, 37, 38, 39, 40, 41]
    right = [42, 43, 44, 45, 46, 47]
    
    #shape = shape_to_np(shape)
    mask_left = np.zeros(face_img.shape[:2], dtype=np.uint8)
    mask_left = eye_on_mask(mask_left, left)
    mask_right = np.zeros(face_img.shape[:2], dtype=np.uint8)
    mask_right = eye_on_mask(mask_right, right)
    
    left_eye = get_eye_bbox(mask_left)
    right_eye = get_eye_bbox(mask_right)

    eyes = [left_eye, right_eye]
    return eyes
```

```python
def shape_to_np(shape, dtype="int"):
    """
    Input:
    shape(object): this is the object returned by dlib.shape_predictor
                    this object contains 68 facial landmarks   
    Output:
    coords(2 numpy array): coordinates of all landmarks
    """
    coords = np.zeros((68, 2), dtype=dtype)
    for i in range(0, 68):
        coords[i] = (shape.part(i).x, shape.part(i).y)
    return coords

def eye_on_mask(mask, side):
    """
    This function converts the eye coordinates into convex mask
    """
    points = [shape[i] for i in side]
    points = np.array(points, dtype=np.int32)
    mask = cv2.fillConvexPoly(mask, points, 255)
    return mask

def get_eye_bbox(mask):
    """
    This function converts masks into eye coordinates
    """
    eye_coord = np.where(mask == 255)
    ys = eye_coord[0]
    xs = eye_coord[1]
    
    left, top, right, bottom = min(xs), min(ys), max(xs), max(ys)
    return left-15, top-25, right+15, bottom+25
```

```python
def predict(model,img_path):
    """
    This function take model and image path and return 
    prediction category and confidence. 
    """
    img_array=cv2.imread(img_path)
    img_resized=cv2.resize(img_array,(200,200))
    
    if h>200 and w>200:
        gray = cv2.cvtColor(img_array, cv2.COLOR_BGR2GRAY)
    else:
        gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)

    face_rect = dlib.rectangle(left=0, top=0, right=w, bottom=h)
    shape = predictor(gray, face_rect)
    shape = shape_to_np(shape)
    eyes = detect_eyes(gray, shape)

    eye_arr=get_eyes_array(img_array, eyes, target_size=(40,40))

    fd, _ = hog(img_resized, orientations=8, pixels_per_cell=(16, 16),
                cells_per_block=(1, 1), visualize=True, channel_axis=-1)

    data_arr=np.concatenate((fd,eye_arr))
    data_arr=data_arr.reshape(1,-1)
    
    probability=model.predict_proba(data_arr)
    predicted_class=model.predict(data_arr)
    predicted_class=predicted_class[0]
    return Categories[predicted_class] , probability[0][predicted_class]
```

```python
#Visualize image and eyes bbox
filepath = '/home/spanidea-168/Documents/SpanIdea_Office_work/proctoring_cropped_dataset/front/img_70.jpg'
img = cv2.imread(filepath)
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
h, w = gray.shape
face_rect = dlib.rectangle(left=0, top=0, right=w, bottom=h)
shape = predictor(gray, face_rect)
shape = shape_to_np(shape)
eyes = detect_eyes(gray, shape)

for (ex1, ey1, ex2, ey2) in eyes:
    cv2.rectangle(img,(ex1,ey1),(ex2, ey2), (0,0,255), 2)

plt.imshow(img)
```

## Prepare data to train SVM

```python
#Prepare data to train SVM

Categories=['front','side']
flat_data_arr=[] #input array
target_arr=[] #output array
datadir='/home/spanidea-168/Documents/SpanIdea_Office_work/proctoring_cropped_dataset' #dataset directory path
#path which contains all the categories of images
for i in Categories:
    
    print(f'loading... category : {i}')
    path=os.path.join(datadir,i)
    for img in os.listdir(path):
        img_array=cv2.imread(os.path.join(path,img))
        h, w, _ = img_array.shape
        #print(img_array.shape)
        #img_resized=cv2.resize(img_array,(150,150))
        img_resized=cv2.resize(img_array,(200,200))
        
        if h>200 and w>200:
            gray = cv2.cvtColor(img_array, cv2.COLOR_BGR2GRAY)
        else:
            gray = cv2.cvtColor(img_resized, cv2.COLOR_BGR2GRAY)

        face_rect = dlib.rectangle(left=0, top=0, right=w, bottom=h)
        shape = predictor(gray, face_rect)
        shape = shape_to_np(shape)
        eyes = detect_eyes(gray, shape)

        eye_arr=get_eyes_array(img_array, eyes, target_size=(40,40))
        
        #img_resized1=cv2.resize(img_array,(70,70))
        #img_array=img_resized1.flatten()
        fd, _ = hog(img_resized, orientations=8, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualize=True, channel_axis=-1)
        
        #eye_arr=get_eyes_array(img_resized)
        data_arr=np.concatenate((fd,eye_arr))
        flat_data_arr.append(data_arr)
        target_arr.append(Categories.index(i))
        
    print(f'loaded category:{i} successfully')
    
flat_data=np.array(flat_data_arr)
target=np.array(target_arr)
df=pd.DataFrame(flat_data) #dataframe
df['Target']=target
x=df.iloc[:,:-1] #input data 
y=df.iloc[:,-1] #output data
```

```python
#Shuffle data

flat_data, target = shuffle(flat_data, target)
```

```python
#Initialize SVM model

param_grid={'C':[0.1,1,10,100],'gamma':[0.0001,0.001,0.1,1],'kernel':['rbf','poly']}
svc=svm.SVC(probability=True)
model=GridSearchCV(svc,param_grid)
```

```python
#Split dataset

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=77,stratify=y)
print('Splitted Successfully')
```

```python
#fit the model

model.fit(x_train,y_train)
print('The Model is trained with the given images')
# model.best_params_ contains the best parameters obtained from GridSearchCV
```

```python
#Evaluate model

y_pred=model.predict(x_test)
print("The predicted Data is :")
print(y_pred)
print("The actual data is:")
print(np.array(y_test))
print(f"The model is {accuracy_score(y_pred,y_test)*100}% accurate")
```

```python
#Predict result

img_path="/home/spanidea-168/Documents/SpanIdea_Office_work/proctoring_cropped_dataset/side/img_350.jpg"
category,probability=predict(model,img_path)
img=imread(img_path)
plt.imshow(img)
plt.show()
img_resize=resize(img,(150,150,3))

# for ind,val in enumerate(Categories):
#     print(f'{val} = {probability[0][ind]*100}%')
print("The predicted image is : "+ category , probability)
```
